```python
%matplotlib inline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn import metrics, datasets
from sklearn.model_selection import KFold
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingClassifier, GradientBoostingRegressor
```

### Regressão (DecisionTree, RandomForest e Gradient Boosting)
* Dataset: boston house-prices dataset
* Aplicar Standardization nas features.
* Comparar os resultados dos seguintes algoritmos: Gradient Descent, Linear Regression, kNN, SVM.
* Usar as seguintes métricas: RMSE, MAE.
* Escolher a melhor técnica e criar um modelo final usando todo o dataset para treinar o modelo.

## Regressão

```python
X, y = datasets.load_boston(return_X_y=True)
```

```python
splits = 5
kf = KFold(n_splits=splits, shuffle=True, random_state=42)
```

```python
# Stardatização dos dados
std_scale = StandardScaler().fit(X)
X = std_scale.transform(X)
```

```python
values_learning_rate = [0.1, 0.05, 0.01]
values_n_estimators = [50, 100, 200]
values_max_depth = [3, 5, 7]
```

#### RandomForest

```python
for i in values_n_estimators:
    for j in values_max_depth:
        rf_rmse = []
        rf_mae = []

        for train_index, test_index in kf.split(X):
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]

            model_rf = RandomForestRegressor(random_state=0, n_estimators=i, max_depth=j)
            model_rf.fit(X_train, y_train)
            y_pred_rf = model_rf.predict(X_test)

            rf_rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred_rf)))
            rf_mae.append(metrics.mean_absolute_error(y_test, y_pred_rf))
        
        print("N_estimators " + str(i) + " ----- Max_depth " + str(j))
        print("RMSE: " + str(np.mean(rf_rmse)))
        print("MAE: " + str(np.mean(rf_mae)))
        print("\n ------------------------- \n")
```

#### GradientBoosting

```python
for i in values_n_estimators:
    for j in values_max_depth:
        for k in values_learning_rate:
            gb_rmse = []
            gb_mae = []

            for train_index, test_index in kf.split(X):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y[train_index], y[test_index]

                model_gb = GradientBoostingRegressor(random_state=0, n_estimators=i, max_depth=j, learning_rate=k)
                model_gb.fit(X_train, y_train)
                y_pred_gb = model_gb.predict(X_test)

                gb_rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred_gb)))
                gb_mae.append(metrics.mean_absolute_error(y_test, y_pred_gb))

            print("N_estimators " + str(i) + " ----- Max_depth " + str(j) + " ----- Learning_rate " + str(k))
            print("RMSE: " + str(np.mean(gb_rmse)))
            print("MAE: " + str(np.mean(gb_mae)))
            print("\n ------------------------- \n")
```

#### DecisionTree

```python
dt_rmse = []
dt_mae = []
for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model_dt = DecisionTreeRegressor(random_state=0)
        model_dt.fit(X_train, y_train)
        y_pred_dt = model_dt.predict(X_test)

        dt_rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred_dt)))
        dt_mae.append(metrics.mean_absolute_error(y_test, y_pred_dt))
    
print("RMSE: " + str(np.mean(dt_rmse)))
print("MAE: " + str(np.mean(dt_mae)))
print("\n ------------------------- \n")
```

## Classificação

```python
X, y = datasets.load_breast_cancer(return_X_y=True)
```

```python
splits = 5
kf = KFold(n_splits=splits, shuffle=True, random_state=42)
```

```python
# Normalização dos dados
norm_scale = MinMaxScaler().fit(X)
X = norm_scale.transform(X)
```

#### RandomForest

```python
for i in values_n_estimators:
    for j in values_max_depth:
        rf_accuracy = []
        rf_precision = []
        rf_recall = []
        rf_f1_measure = []
        rf_auc_ = []
        rf_fpr_ = []
        rf_tpr_ = []
        
        for train_index, test_index in kf.split(X):
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]

            model_rf = RandomForestClassifier(random_state=0, n_estimators=i, max_depth=j)
            model_rf.fit(X_train, y_train)
            y_pred_rf = model_rf.predict(X_test)

            rf_accuracy.append(metrics.accuracy_score(y_test, y_pred_rf))
            rf_precision.append(metrics.precision_score(y_test, y_pred_rf))
            rf_recall.append(metrics.recall_score(y_test, y_pred_rf))
            rf_f1_measure.append(metrics.f1_score(y_test, y_pred_rf))
            rf_auc_.append(metrics.roc_auc_score(y_test, y_pred_rf))
            fpr, tpr, thr = metrics.roc_curve(y_test, y_pred_rf)
            rf_fpr_.append(fpr[1])
            rf_tpr_.append(tpr[1])
            
        
        print("N_estimators " + str(i) + " ----- Max_depth " + str(j))
        print("Accuracy: " + str(np.mean(rf_accuracy)))
        print("Precision: " + str(np.mean(rf_precision)))
        print("Recall: " + str(np.mean(rf_recall)))
        print("F1: " + str(np.mean(rf_f1_measure)))
        print("AUC: " + str(np.mean(rf_auc_)))
        print(metrics.confusion_matrix(y_test, y_pred_rf, labels=np.unique(y_test)))
        
        
        roc = metrics.auc(fpr, tpr)
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange',
                 lw=2, label='ROC curve (area = %0.2f)' % roc)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve')
        plt.legend(loc="lower right")
        plt.show()
        print("\n ------------------------- \n")
        
```

#### GradientBoosting

```python
for i in values_n_estimators:
    for j in values_max_depth:
        for k in values_learning_rate:
            gb_accuracy = []
            gb_precision = []
            gb_recall = []
            gb_f1_measure = []
            gb_auc_ = []
            gb_fpr_ = []
            gb_tpr_ = []

            for train_index, test_index in kf.split(X):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y[train_index], y[test_index]

                model_gb = GradientBoostingClassifier(random_state=0, n_estimators=i, max_depth=j, learning_rate=k)
                model_gb.fit(X_train, y_train)
                y_pred_gb = model_gb.predict(X_test)

                gb_accuracy.append(metrics.accuracy_score(y_test, y_pred_gb))
                gb_precision.append(metrics.precision_score(y_test, y_pred_gb))
                gb_recall.append(metrics.recall_score(y_test, y_pred_gb))
                gb_f1_measure.append(metrics.f1_score(y_test, y_pred_gb))
                gb_auc_.append(metrics.roc_auc_score(y_test, y_pred_gb))
                fpr, tpr, thr = metrics.roc_curve(y_test, y_pred_gb)
                gb_fpr_.append(fpr[1])
                gb_tpr_.append(tpr[1])

            print("N_estimators " + str(i) + " ----- Max_depth " + str(j) + " ----- Learning_rate " + str(k))
            print("Accuracy: " + str(np.mean(gb_accuracy)))
            print("Precision: " + str(np.mean(gb_precision)))
            print("Recall: " + str(np.mean(gb_recall)))
            print("F1: " + str(np.mean(gb_f1_measure)))
            print("AUC: " + str(np.mean(gb_auc_)))
            print(metrics.confusion_matrix(y_test, y_pred_gb, labels=np.unique(y_test)))

            roc = metrics.auc(fpr, tpr)
            plt.figure()
            plt.plot(fpr, tpr, color='darkorange',
                     lw=2, label='ROC curve (area = %0.2f)' % roc)
            plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
            plt.xlim([0.0, 1.0])
            plt.ylim([0.0, 1.05])
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title('ROC Curve')
            plt.legend(loc="lower right")
            plt.show()
            print("\n ------------------------- \n")
        
```

#### DecisionTree

```python
dt_accuracy = []
dt_precision = []
dt_recall = []
dt_f1_measure = []
dt_auc_ = []
dt_fpr_ = []
dt_tpr_ = []

for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model_dt = DecisionTreeClassifier(random_state=0)
        model_dt.fit(X_train, y_train)
        y_pred_dt = model_dt.predict(X_test)

        dt_accuracy.append(metrics.accuracy_score(y_test, y_pred_dt))
        dt_precision.append(metrics.precision_score(y_test, y_pred_dt))
        dt_recall.append(metrics.recall_score(y_test, y_pred_dt))
        dt_f1_measure.append(metrics.f1_score(y_test, y_pred_dt))
        dt_auc_.append(metrics.roc_auc_score(y_test, y_pred_dt))
        fpr, tpr, thr = metrics.roc_curve(y_test, y_pred_dt)
        dt_fpr_.append(fpr[1])
        dt_tpr_.append(tpr[1])
        
print("Accuracy: " + str(np.mean(dt_accuracy)))
print("Precision: " + str(np.mean(dt_precision)))
print("Recall: " + str(np.mean(dt_recall)))
print("F1: " + str(np.mean(dt_f1_measure)))
print("AUC: " + str(np.mean(dt_auc_)))
print(metrics.confusion_matrix(y_test, y_pred_dt, labels=np.unique(y_test)))

roc = metrics.auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange',
        lw=2, label='ROC curve (area = %0.2f)' % roc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
print("\n ------------------------- \n")
        
```

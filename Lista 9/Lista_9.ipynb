#### Missing Values (NaN)
* As features que possuem valores nulos são:
    - (1) Workclass
    - (6) Occupation
    - (13) Native-Country
    
* As colunas serão modificadas da seguinte forma: Essas colunas serão modificadas baseando-se na média

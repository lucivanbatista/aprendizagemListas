```python
%matplotlib inline
from sklearn import preprocessing
from sklearn import metrics, linear_model, datasets, neighbors, svm
from sklearn.model_selection import KFold
from sklearn import naive_bayes
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
```

```python
bos = bos.drop(columns = ['CHAS', 'RAD'])
```

```python
cor = bos.corr()
cor
```

```python
#X, y = datasets.load_boston(return_X_y=True)
X = bos.as_matrix(columns=['DIS', 'B', 'CRIM', 'AGE', 'INDUS', 'NOX','RM','TAX','PTRATIO','LSTAT', 'ZN'])
y = bos.as_matrix(columns=['Price'])[:,0]
```

```python
print(X.shape)
print(y.shape)
```

```python
splits = 5
kf = KFold(n_splits=splits, shuffle=True, random_state=42)
kf.get_n_splits(X, y)
```

```python
sgd_rmse = []
sgd_mae = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    std_scale = preprocessing.StandardScaler().fit(X_train)
    X_train = std_scale.transform(X_train)
    X_test = std_scale.transform(X_test)
    
    model_sgd = linear_model.SGDRegressor(max_iter=10)
    model_sgd.fit(X_train, y_train)
    y_pred_sgd = model_sgd.predict(X_test)
    
    sgd_rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred_sgd)))
    sgd_mae.append(metrics.mean_absolute_error(y_test, y_pred_sgd))
```

```python
lr_rmse = []
lr_mae = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    std_scale = preprocessing.StandardScaler().fit(X_train)
    X_train = std_scale.transform(X_train)
    X_test = std_scale.transform(X_test)
    
    model_lr = linear_model.LinearRegression()
    model_lr.fit(X_train, y_train)
    y_pred_lr = model_lr.predict(X_test)
    
    lr_rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred_lr)))
    lr_mae.append(metrics.mean_absolute_error(y_test, y_pred_lr))
```

```python
knn_rmse = []
knn_mae = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    std_scale = preprocessing.StandardScaler().fit(X_train)
    X_train = std_scale.transform(X_train)
    X_test = std_scale.transform(X_test)
    
    model_knn = neighbors.KNeighborsRegressor()
    model_knn.fit(X_train, y_train)
    y_pred_knn = model_knn.predict(X_test)
    
    knn_rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred_knn)))
    knn_mae.append(metrics.mean_absolute_error(y_test, y_pred_knn))
```

```python
svr_rmse = []
svr_mae = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    std_scale = preprocessing.StandardScaler().fit(X_train)
    X_train = std_scale.transform(X_train)
    X_test = std_scale.transform(X_test)
    
    model_svr = svm.SVR()
    model_svr.fit(X_train, y_train)
    y_pred_svr = model_svr.predict(X_test)
    
    svr_rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred_svr)))
    svr_mae.append(metrics.mean_absolute_error(y_test, y_pred_svr))
```

```python
print('RMSE: ')
print('--- SGD: ' + str(np.mean(sgd_rmse)))
print('--- LR: ' + str(np.mean(lr_rmse)))
print('--- KNN: ' + str(np.mean(knn_rmse)))
print('--- SVR: ' + str(np.mean(svr_rmse)))

print('\nMAE: ')
print('--- SGD: ' + str(np.mean(sgd_mae)))
print('--- LR: ' + str(np.mean(lr_mae)))
print('--- KNN: ' + str(np.mean(knn_mae)))
print('--- SVR: ' + str(np.mean(svr_mae)))
```

#### Resposta: O menor RMSE e o menor MAE pertece ao KNN

```python
#Modelo final
std_scale = preprocessing.StandardScaler().fit(X)
X = std_scale.transform(X)

model = neighbors.KNeighborsRegressor(n_neighbors=2)
model.fit(X, y)
y_pred = model.predict(X)
rmse = math.sqrt(metrics.mean_squared_error(y, y_pred))
mae = metrics.mean_absolute_error(y, y_pred)
print('RMSE Final: ' + str(rmse))
print('MAE Final: ' + str(mae))
```

## Classificação

```python
X, y = datasets.load_breast_cancer(return_X_y=True)
print(X.shape)
print(y.shape)
```

```python
norm_scale = preprocessing.MinMaxScaler()
norm_scale.fit(X)
X_norm = norm_scale.transform(X)
```

```python
kf = KFold(n_splits=splits, shuffle=True)
kf.get_n_splits(X_norm)
```

```python
logr_accuracy = []
logr_precision = []
logr_recall = []
logr_f1_measure = []
logr_auc_ = []


for train_index, test_index in kf.split(X_norm):
    X_train, X_test = X_norm[train_index], X_norm[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    model_logr = linear_model.LogisticRegression()
    model_logr.fit(X_train, y_train)
    y_pred_logr = model_logr.predict(X_test)
    
    logr_accuracy.append(metrics.accuracy_score(y_test, y_pred_logr))
    logr_precision.append(metrics.precision_score(y_test, y_pred_logr))
    logr_recall.append(metrics.recall_score(y_test, y_pred_logr))
    logr_f1_measure.append(metrics.f1_score(y_test, y_pred_logr))
    
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_logr)
    logr_auc = metrics.auc(fpr, tpr)
    logr_auc_.append(logr_auc)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange',
             lw=2, label='ROC curve (area = %0.2f)' % logr_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
knn_accuracy = []
knn_precision = []
knn_recall = []
knn_f1_measure = []
knn_auc_ = []

for train_index, test_index in kf.split(X_norm):
    X_train, X_test = X_norm[train_index], X_norm[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    model_knn = neighbors.KNeighborsClassifier()
    model_knn.fit(X_train, y_train)
    y_pred_knn = model_knn.predict(X_test)
    
    knn_accuracy.append(metrics.accuracy_score(y_test, y_pred_knn))
    knn_precision.append(metrics.precision_score(y_test, y_pred_knn))
    knn_recall.append(metrics.recall_score(y_test, y_pred_knn))
    knn_f1_measure.append(metrics.f1_score(y_test, y_pred_knn))
    
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_knn)
    knn_auc = metrics.auc(fpr, tpr)
    knn_auc_.append(knn_auc)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange',
             lw=2, label='ROC curve (area = %0.2f)' % knn_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
nb_accuracy = []
nb_precision = []
nb_recall = []
nb_f1_measure = []
nb_auc_ = []

for train_index, test_index in kf.split(X_norm):
    X_train, X_test = X_norm[train_index], X_norm[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    model_nb = naive_bayes.GaussianNB()
    model_nb.fit(X_train, y_train)
    y_pred_nb = model_knn.predict(X_test)
    
    nb_accuracy.append(metrics.accuracy_score(y_test, y_pred_nb))
    nb_precision.append(metrics.precision_score(y_test, y_pred_nb))
    nb_recall.append(metrics.recall_score(y_test, y_pred_nb))
    nb_f1_measure.append(metrics.f1_score(y_test, y_pred_nb))
    
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_nb)
    nb_auc = metrics.auc(fpr, tpr)
    nb_auc_.append(nb_auc)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange',
             lw=2, label='ROC curve (area = %0.2f)' % nb_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
svc_accuracy = []
svc_precision = []
svc_recall = []
svc_f1_measure = []
svc_auc_ = []

for train_index, test_index in kf.split(X_norm):
    X_train, X_test = X_norm[train_index], X_norm[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    
    model_svc = svm.SVC()
    model_svc.fit(X_train, y_train)
    y_pred_svc = model_svc.predict(X_test)
    
    
    svc_accuracy.append(metrics.accuracy_score(y_test, y_pred_svc))
    svc_precision.append(metrics.precision_score(y_test, y_pred_svc))
    svc_recall.append(metrics.recall_score(y_test, y_pred_svc))
    svc_f1_measure.append(metrics.f1_score(y_test, y_pred_svc))
    
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_svc)
    svc_auc = metrics.auc(fpr, tpr)
    svc_auc_.append(svc_auc)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange',
             lw=2, label='ROC curve (area = %0.2f)' % svc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

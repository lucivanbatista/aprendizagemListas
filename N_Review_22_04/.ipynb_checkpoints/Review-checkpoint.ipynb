#### Explicação brévia:
* Mean representa a média dos valores do array (valor do meio de uma distribuição normal)
* Std ou desvio padrão representa o quão os pontos estão afastados da média
* Variancia indica o quão longe os valores estão do valor esperado

* MSE (Erro quadrado médio)
* RMSE normalmente representa um valor "margem de erro" para as predições
* MAE 

```python
#Usando o dataset de points
# mean; std; var;
print('Mean: ' + str(np.mean(X)))
print('Std: ' + str(np.std(X)))
print('Var: ' + str(np.var(X)))

m = linear_model.LinearRegression()
m.fit(X, y)
y_pred = m.predict(X)

#MSE; RSME; MAE
print('\nMSE: ' + str(metrics.mean_squared_error(y, y_pred)))
print('RMSE: ' + str(math.sqrt(metrics.mean_squared_error(y, y_pred))))
print('MAE: ' + str(metrics.mean_absolute_error(y, y_pred)))
```

```python
from sklearn.model_selection import KFold
```

```python
train = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
train.get_n_splits(X, y)
```

```python
rmse = []
mae = []
for train_index, test_index in train.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    std_scale = preprocessing.StandardScaler().fit(X_train)
    X_train_std = std_scale.transform(X_train)
    X_test_std = std_scale.transform(X_test)
    
    model = linear_model.LinearRegression()
    model.fit(X_train_std, y_train)
    
    y_pred = model.predict(X_test_std)
    rmse.append(math.sqrt(metrics.mean_squared_error(y_test,y_pred)))
    mae.append(metrics.mean_absolute_error(y_test,y_pred))
    
print(rmse)
print(mae)

print(np.mean(rmse))
print(np.mean(mae))
```

### Lista 7

```python
from sklearn import datasets
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
X.shape
```

```python
kfold = KFold(n_splits=5)
kfold.get_n_splits(X, y)
```

```python
# Using Linear Regression
rmse = []
mae = []

for idx_train, idx_test in kfold.split(X, y):
    X_train, X_test = X[idx_train], X[idx_test]
    y_train, y_test = y[idx_train], y[idx_test]
    
    scale_std = preprocessing.StandardScaler().fit(X_train)
    X_train_std = scale_std.transform(X_train)
    X_test_std = scale_std.transform(X_test)
    
    model = linear_model.LinearRegression()
    model.fit(X_train_std, y_train)
    
    y_pred = model.predict(X_test_std)
    
    rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    mae.append(metrics.mean_absolute_error(y_test, y_pred))
    
print(rmse)
print(mae)
print(np.mean(rmse))
print(np.mean(mae))
```

```python
from sklearn import neighbors
```

```python
# Using KNN
rmse = []
mse = []

for idx_train, idx_test in kfold.split(X, y):
    X_train, X_test = X[idx_train], X[idx_test]
    y_train, y_test = y[idx_train], y[idx_test]
    
    scale_std = preprocessing.StandardScaler().fit(X_train)
    X_train_std = scale_std.transform(X_train)
    X_test_std = scale_std.transform(X_test)
    
    model = neighbors.KNeighborsRegressor()
    model.fit(X_train_std, y_train)
    y_pred = model.predict(X_test_std)
    
    rmse.append(math.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    mae.append(metrics.mean_absolute_error(y_test, y_pred))
    
print(rmse)
print(mae)
print(np.mean(rmse))
print(np.mean(mae))

# Também pode ser usado o SGDRegression
# Tabmém pode ser usado o SVM (Regression)
```
